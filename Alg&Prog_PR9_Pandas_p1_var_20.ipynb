##Задача 3. Работа со строковыми данными в `pandas`

### 3.1 Измените название каждой марки автомобиля в таблице `vehicles` таким образом, чтобы каждое слово в названии начиналось с прописной буквы.

````python
vehicles.brand = vehicles.brand.str.title()
vehicles
````

### 3.2 Добавьте в таблицу `vehicles` столбец `name_word_count`, в котором хранится количество слов из названия марки (считайте, что слова в названии разделяются только пробелами). Обратите внимание, что между словами может располагаться несколько пробелов подряд.

````python
vehicles = vehicles.assign(name_word_count=vehicles.brand.str.split().str.len())
vehicles
````

### 3.3 Добавьте в таблицу vehicles столбец description_word_count, в котором хранится количество слов в описании автомобиля (считайте, что слова в описании разделяются только пробелами).

````python
vehicles = vehicles.assign(description_word_count=vehicles.description.str.split().str.len())
vehicles
````

### 3.4 Отсортируйте таблицу в порядке убывания величины столбца `name_word_count` и сохраните ее в csv файл.

````python
vehicles.sort_values('name_word_count', ascending=False).to_csv('./sorted_vehicles.csv')
````
